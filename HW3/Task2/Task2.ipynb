```python
import numpy as np
import pandas as pd
import pylab as pl
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import export_graphviz
import pydot 
import subprocess
```

# Titanic

```python
df=pd.read_csv(r'D:\DataAnalysysHW\HW3\Dataset\titanic.csv',encoding="cp1251")
print(df.shape) 
print(df.info())
```

```python
df.describe()
```

```python
x=df[['Survived','Pclass','Age','Siblings/Spouses Aboard','Parents/Children Aboard','Fare']]
y=df.values[::,7]
y=y.astype('int')
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
tree=DecisionTreeClassifier(max_depth=5, random_state=0)
tree.fit(x_train, y_train)
```

```python
export_graphviz(tree, out_file='Titanic_tree.dot', 
rounded = True, proportion = False, 
                precision = 2, filled = True)

(graph, ) = pydot.graph_from_dot_file('Titanic_tree.dot')

graph.write_png('Titanic_tree.png')
```

```python
tree.score(x_test,y_test)
```

```python
tree.score(x_train, y_train)
```
